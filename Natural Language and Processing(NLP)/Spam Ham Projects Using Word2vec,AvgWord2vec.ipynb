```python
import pandas as pd
messages=pd.read_csv('Spam_Classifier\sms_spam_collection\SMSSpamCollection',
                    sep='\t',names=["label","message"])
```

```python
messages.shape
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    
    review = [lemmatizer.lemmatize(word) for word in review]
    review = ' '.join(review)
    corpus.append(review)
```

```python
[[i,j,k] for i,j,k in zip(list(map(len,corpus)),corpus, messages['message']) if i<1]
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words=[]
for sent in corpus:
    sent_token=sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
import numpy as np

# `embedding_size` is the dimensionality of each Word2Vec vector (e.g., 100)
embedding_size = 100  

# Ensure each entry in X has a fixed length by averaging word embeddings
X_new = np.array([
    np.mean(vec, axis=0) if isinstance(vec, list) and len(vec) > 0 else vec
    if isinstance(vec, np.ndarray) and vec.shape[0] == embedding_size
    else np.zeros(embedding_size)
    for vec in X
])

""" For Lists of Embeddings: If vec is a non-empty list (e.g., multiple word vectors for a text), it averages them to create a single fixed-length vector.
For Single Vectors: If vec is a single vector with a length of embedding_size, it uses it directly.
For Empty or Invalid Entries: If neither of the above conditions is met (e.g., no valid words in the text), it assigns a zero vector of length embedding_size.
"""
print(X_new.shape)  # Should print (5569, embedding_size)
```
